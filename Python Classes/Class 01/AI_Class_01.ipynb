**Calculator**

```
def add(a,b):
  return a+b
def sub(a,b):
  return a-b
def mul(a,b):
  return a * b
def div(a,b):
  try:
    return a / b
  except ZeroDivisionError:
    print('Cannot divide by zero')

def calculator(a,b):
  print("Enter '1' for add,\nEnter '2' for sub,\nEnter '3' for mul,\nEnter '4' for div")
  chose = int(input('Enter your choice: '))
  if chose == 1:
    print(add(a,b))
  elif chose == 2:
    print(sub(a,b))
  elif chose == 3:
    print(mul(a,b))
  elif chose == 4:
    print(div(a,b))
  else:
    print('Enter the given number')
qw1 = float(input('Enter the first number: '))
qw2 = float(input('Enter the second number: '))
cal = calculator(qw1, qw2)
```

**Percentage Calculator**

```
def user_input_marks(c):
    marks = []
    for i in range(c):
        b = int(input(f'Enter marks for subject {i}: '))
        if 0 <= b <= 100:
            marks.append(b)
        else:
            print("Marks between 0 and 100")
    return marks
def percentage_calculator(c):
    marks = user_input_marks(c)
    total_marks = len(marks) * 100
    obtained_marks = sum(marks)
    if total_marks == 0:
        return 0
    return (obtained_marks / total_marks) * 100
percent = percentage_calculator(5)
print("Percentage:", percent)
```

**Grade System**

```
def grade_system():
  try:
    grade = int(input('Enter the number: '))
    if grade > 80:
      return 'A'
    elif grade > 70:
      return 'B'
    elif grade > 55:
      return 'C'
    elif grade > 45:
      return 'D'
    else:
      return 'F'
  except ValueError:
    print("Enter correct data type")
    return 0
print(grade_system())
```
