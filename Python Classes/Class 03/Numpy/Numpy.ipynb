**Linear Algebra**

Dot Product

```python
q = np.array([
    [1,2],
    [3,4]
])
p = np.array([
    [5,6],
    [7,8]
])
print(np.dot(q,p))
```

```python
q1 = np.array([
    [2,0],
    [1,3]
])
p1 = np.array([
    [1,2],
    [3,4]
])

d1 = np.dot(q1,p1)
print(d1.shape)
```

```python
s3 = d1.reshape(1,2,2)
w3 = np.arange(8).reshape(2,2,2)
print(np.dot(s3,w3))
```

```python
s1 = np.random.randint(1, 10, size=8).reshape(2,2,2)
s2 = np.random.randint(1, 5, size=4).reshape(1,2,2)
print('Dot product:\n\n',np.dot(s1,s2))
```

**Iteration**

```python
k = np.arange(6).reshape(2,3)
print(k)
for i in k:
  print(i)
for i in np.nditer(k):
  print(i)
```

**Transpose**

```python
print(k.T)
```

**Ravel:**
The numpy.ravel() function is used to flatten a multi-dimensional NumPy array into a one-dimensional array.

```python
print(np.ravel(k))
```

**Stacking**

numpy.vstack(tup): Stacks arrays vertically (row-wise), which is equivalent to concatenation along the first axis (axis=0). This is useful for combining arrays with the same number of columns.

```python
a4 = np.array([[1, 2], [3, 4]])
b4 = np.array([[5, 6], [7, 8]])
result = np.vstack([a4, b4])
print(result)
```

numpy.hstack(tup): Stacks arrays horizontally (column-wise), equivalent to concatenation along the second axis (axis=1) for 2D arrays. This is useful for combining arrays with the same number of rows.

```python
a5 = np.array([[1, 2], [3, 4]])
b5 = np.array([[5, 6], [7, 8]])
result = np.hstack([a5, b5])
print(result)
```
