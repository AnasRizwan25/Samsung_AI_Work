What a Percentile?

The n-th percentile is the value below which n% of the data falls.

Example: the 75th percentile is the value where 75% of the data is less than or equal to it.

```python
np.sort(marks)
```

```python
percentile_75 = np.percentile(marks, 75)
percentile_50 = np.percentile(marks, 50)
percentile_25 = np.percentile(marks, 25)
percentile_25
percentile_50
percentile_75
```
