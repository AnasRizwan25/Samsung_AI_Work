```python
import numpy as np
```

```python
data2 = {
    'name' : ['anas', 'talha', 'zaid'],
    'mark' : [76, 96, np.nan]
}
df = pd.DataFrame(data2)
df
```

```python
# Detect Missing
df.isnull()
```

```python
df.isnull().sum() # give the sum of null value of given column
```

```python
df.dropna() # Removes rows containing missing values.
```

```python
df.fillna(0) # Fill missing data
df.loc[0, 'mark'] = np.nan
```

```python
df['mark'] = df['mark'].fillna(df['mark'].mean())
```
