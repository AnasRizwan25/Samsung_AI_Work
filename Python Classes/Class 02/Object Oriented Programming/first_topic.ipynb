**Object Oriented Programming:**
Object-Oriented Programming is a programming paradigm where we design software using objects instead of just functions.

An object combines data (attributes) and behavior (methods) into a single unit and models real-world entities like user, account, car, etc.

Q: Why do we use OOP?

We use Object-Oriented Programming to manage complex software more effectively by organizing code around real-world entities instead of long procedural logic.
It improves structure, scalability, and maintainability of applications.

*   __init__: A initialize value to the object value.
*   **self**: Refer the current object.

| Procedural Oriented Programming | Object-Oriented Programming                 |
| ------------------------------------- | -------------------------------------------------- |
| Program divided into functions        | Program divided into objects                       |
| Follows top-down approach             | Follows bottom-up approach                         |
| No access specifiers                  | Has access specifiers (private, public, protected) |
| Adding new data/functions difficult   | Adding new data/functions easy                     |
| No proper data hiding → less secure   | Provides data hiding → more secure                 |
| Overloading not possible              | Overloading possible                               |
| No inheritance concept                | Supports inheritance                               |
| Functions more important than data    | Data more important than functions                 |
| Based on unreal world approach        | Based on real-world modeling                       |
| Used for medium-sized programs        | Used for large and complex programs                |
| Uses procedural abstraction           | Uses data abstraction                              |
| Code reusability absent               | Code reusability present                           |
| Examples: C, FORTRAN, Pascal, BASIC   | Examples: C++, Java, Python, C#                    |





```python
class student:
  def __init__(self, name, mark):
    self.name = name
    self.mark = mark
  def get_student_info(self):
    return f'The student name: {self.name} and marks: {self.mark}'
try:
  s1 = student('Ali', 70)
  print(s1.get_student_info())
except Exception as e:
  print(e)
```

**Encapsulation**

```python
class BankAccount:
  def __init__(self, balance):
    self.__balance = balance
  def deposit(self, amount):
    if amount == 0:
      return 0
    self.__balance += amount
  def get_balance(self):
    return self.__balance

try:
  bank_a = BankAccount(2000)
  bank_a.deposit(2000)
  print(bank_a.get_balance())
except Exception as e:
  print(e)
```

**In this exercise, you'll start building the Employee class, beginning by defining it as an empty class, and then adding a method allowing you to set an object's .name attribute**.

```python
class Employee:
  # Include a set_name method
  def set_name(self, set_name):
    self.name = set_name

emp = Employee()

# Use set_name() on emp to set the name of emp to 'Korel Rossi'
emp.set_name('Korel Rossi')
print(emp.name)
```

```python
class Employee:
    min_salary = 30000   # class variable
    def __init__(self, name, salary=0):
        self.name = name
        if salary > Employee.min_salary:
            self.salary = salary
        else:
            self.salary = Employee.min_salary
    def info(self):
        return f"Name: {self.name}\nsalary: {self.salary}"


emp = Employee('Korel Rossi', 30000)
print(emp.info())
```

**Class Method**

```python
class Person:
  CURRENT_YEAR = 2024
  def __init__(self, name, age):
    self.name = name
    self.age = age
  
  # Add a class method decorator
  @classmethod
  # Define the from_birth_year method
  def from_birth_year(cls, name, birth_year):
    # Create age
    age = cls.CURRENT_YEAR - birth_year
    # Return the name and age
    return cls(name, age)

bob = Person.from_birth_year("Bob", 1990)
```
